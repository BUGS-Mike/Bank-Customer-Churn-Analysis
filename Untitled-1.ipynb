1.Libraries Import 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset Load

```python
df = pd.read_csv("creditcard.csv")
print(df.head())
```

Basic Dataset Info

```python
print(df.shape)
print(df.info())
```

Check Missing Values

```python
print(df.isnull().sum())
```

Fraud vs Normal Transactions

```python
print(df['Class'].value_counts())
```

Fraud Distribution Plot

```python
sns.countplot(x='Class', data=df)

plt.title("Fraud vs Normal Transactions")
plt.xlabel("Transaction Type")
plt.ylabel("Count")

plt.show()
```

Transaction Amount Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")
plt.xlabel("Amount")
plt.ylabel("Frequency")

plt.show()
```

Fraud Transactions Amount

```python
fraud = df[df['Class'] == 1]

plt.figure(figsize=(8,5))
sns.histplot(fraud['Amount'], bins=30)

plt.title("Fraud Transaction Amounts")
plt.show()
```

A credit card transaction dataset was analyzed to identify patterns in fraudulent activity. Exploratory analysis helped understand transaction behavior and detect irregular patterns associated with fraud. The study highlights how data analysis supports financial institutions in monitoring risks and improving fraud detection systems.

**Key Outcomes**

* Identified distribution of normal vs fraudulent transactions
* Analyzed transaction amount patterns
* Detected irregular transaction behavior
* Improved understanding of fraud trends
* Demonstrated the role of data analytics in financial risk management
