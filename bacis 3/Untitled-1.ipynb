Libraries Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load Dataset

```python
df = pd.read_csv("churn_modelling.csv")
print(df.head())
```

Dataset Info

```python
print(df.shape)
print(df.info())
```

Churn Distribution

```python
sns.countplot(x="Exited", data=df)

plt.title("Customer Churn Distribution")
plt.xlabel("Churn (1 = Left, 0 = Stayed)")
plt.ylabel("Count")

plt.show()
```

Age vs Churn Analysis

```python
plt.figure(figsize=(8,5))

sns.boxplot(x="Exited", y="Age", data=df)

plt.title("Age vs Churn")

plt.show()
```

Balance Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df["Balance"], bins=30)

plt.title("Balance Distribution")

plt.show()
```

Products vs Churn

```python
sns.countplot(x="NumOfProducts", hue="Exited", data=df)

plt.title("Products vs Churn")

plt.show()
```

 Summary :

A bank customer churn dataset was analyzed to understand factors influencing customer attrition. Exploratory data analysis examined relationships between churn and variables such as age, balance, and number of products. The insights help financial institutions identify at-risk customers and develop strategies to improve retention and customer satisfaction.
