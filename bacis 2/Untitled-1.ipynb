Libraries Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Dataset Load

```python
df = pd.read_csv("BankChurners.csv")
print(df.head())
```

Basic Dataset Info

```python
print(df.shape)
print(df.info())
```

Missing Values Check

```python
print(df.isnull().sum())
```

Credit Limit Analysis

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Credit_Limit'], bins=30)

plt.title("Credit Limit Distribution")
plt.xlabel("Credit Limit")
plt.ylabel("Frequency")

plt.show()
```

Customer Age Distribution

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Customer_Age'], bins=30)

plt.title("Customer Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")

plt.show()
```

Average Spending by Card Category

```python
print(df.columns)
```

```python
category_spending = df.groupby("Card_Category")["Total_Trans_Amt"].mean()

print(category_spending)
```

Category Spending Chart

```python
sns.barplot(x=category_spending.index, y=category_spending.values)

plt.title("Average Spending by Card Category")
plt.xlabel("Card Category")
plt.ylabel("Average Transaction Amount")

plt.show()
```

 Summary :

Exploratory analysis revealed patterns in age distribution, transaction amounts, and card category usage.
